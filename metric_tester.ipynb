```python
from trainer import pl_data_module, pl_model, CLASSES, word2idx, idx2word
```

```python
model = pl_model.load_from_checkpoint("/home/captain-america/external_disk/.eff/AttentiveMobileWord-Trainer/results_dir/epoch=61-train_loss_epoch=7.4645.ckpt", class_count=len(CLASSES))
```

```python
pl_data_module.setup(32)
iterator = iter(pl_data_module.val_dataloader())
```

```python
test_entry = next(iterator)
x, y = test_entry['x'], test_entry['y'].squeeze()
y_pred = model(x)

y_pred_armax = y_pred.argmax(axis=0)
for y_pred_i, y_true_i in zip(y_pred_armax, y) :
    print(idx2word[y_pred_i.item()], idx2word[y_true_i.item()])
```

```python
k = 1750
topk = y_pred.topk(k).indices
val = 0.0
for i in range(k):
    row = topk[:, i]
    #print(row.shape)
    accuracy = (row==y).float().mean().item()
    val += accuracy
    #print(accuracy, val)
```
